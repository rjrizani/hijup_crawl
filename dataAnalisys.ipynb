```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('hijup\output.csv')

# Display the first few rows of the dataframe
df.head()
```

```python
df.tail(10)
```

```python
import re

def clean_price(price):
    # Menghapus semua karakter kecuali angka
    return int(re.sub(r'\D', '', price))

# Terapkan fungsi ke kolom 'price'
df['price'] = df['price'].apply(clean_price)

# Tampilkan dataframe untuk memastikan perubahan
df.head()
```

```python
df.describe(include='all')
```

```python
import re

def get_last_text_after_number(url):
    match = re.search(r'\d+-(\w+)$', url)
    return match.group(1) if match else None

df['last_text'] = df['url'].apply(get_last_text_after_number)
df[['url', 'last_text']].head()
```

```python
def get_product_name_from_url(url):
    match = re.search(r'/(\d+)-(.+)$', url)
    return match.group(2).replace('-', ' ') if match else None

# Example usage
url = 'https://www.hijup.com/id/pants/115885-highwaist-jeans-celana-kulot-denim'
product_name = get_product_name_from_url(url)
print(product_name)
```

```python
df['last_text'] = df['url'].apply(get_product_name_from_url)
```
