```python
import matplotlib.pyplot as plt
if df.empty:
    print("No data in df")
else:
    # Plot a simple graph of price_RP
    plt.figure(figsize=(10, 6))
    plt.plot(df['price_RP'])
    plt.title('Price Distribution')
    plt.xlabel('Index')
    plt.ylabel('Price (RP)')
    plt.show()
```

```python

# Check if df is not empty
if df.empty:
    print("No data in df")
else:
    # Group by brand and calculate the mean price for each brand
    brand_price = df.groupby('brand')['price_RP'].mean().sort_values()

    # Check if brand_price is not empty
    if brand_price.empty:
        print("No data in brand_price")
    else:
        # Plot the data
        plt.figure(figsize=(10, 6))
        brand_price.plot(kind='bar')
        plt.title('Average Price by Brand')
        plt.xlabel('Brand')
        plt.ylabel('Average Price (RP)')
        plt.xticks(rotation=45)
        plt.show()
```

```python
# Get the brand with the highest price
highest_price_brand = brand_price.idxmax()

# Get the material for the brand with the highest price
material_highest_price_brand = df[df['brand'] == highest_price_brand]['material'].unique()

print(f"The material for the brand with the highest price ({highest_price_brand}) is: {material_highest_price_brand}")
```
